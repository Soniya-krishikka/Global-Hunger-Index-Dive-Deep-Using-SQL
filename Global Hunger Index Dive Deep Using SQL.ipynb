```python
# import sqlalchemy and create a sqlite engine
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)

# export the dataframe as a table 'playstore' to the sqlite engine
df.to_sql("GHI", con =engine)
```

```python
# Select all columns from the GHI table
sql='''
SELECT * FROM GHI
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head(10)
```

```python
# Select distinct countries from the GHI table
sql='''
SELECT DISTINCT Country FROM GHI
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Count the number of distinct countries in the GHI table
sql='''
SELECT COUNT(DISTINCT Country) AS num_countries FROM GHI
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select countries with a 2023 value GREATER than 10
sql='''
SELECT Country FROM GHI
WHERE "2023" > 10
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select countries and their 2023 values
sql='''
SELECT Country, "2023" FROM GHI
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Select the average 2023 value for all countries
sql='''
SELECT AVG("2023") AS avg_2023 FROM GHI
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select the top 5 countries with the highest 2023 values
sql='''
SELECT Country, "2023" FROM GHI
ORDER BY "2023" DESC
LIMIT 5
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select countries where the % change since 2015 is greater than 20%
sql='''
SELECT Country FROM GHI
WHERE "% change since 2015" > 20
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select countries and their absolute change since 2015
sql='''
SELECT Country, "Absolute change since 2015" FROM GHI
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select the average absolute change since 2015 for all countries
sql='''
SELECT AVG("Absolute change since 2015") AS avg_abs_change FROM GHI
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select the top 3 countries with the highest % change since 2015
sql='''
SELECT Country, "% change since 2015" FROM GHI
ORDER BY "% change since 2015" DESC
LIMIT 3
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select countries that had a decrease in 2023 compared to 2015
sql='''
SELECT Country FROM GHI
WHERE "2023" < "2015"
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select the number of countries with a 2023 value greater than 20
sql='''
SELECT COUNT(*) AS num_countries_above_20 FROM GHI
WHERE "2023" > 20
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select countries with the highest improvement from 2015 to 2023
sql='''
SELECT Country, ("2015" - "2023") AS improvement
FROM GHI
ORDER BY improvement DESC
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Calculate the weighted average 2023 values for each rank, weighted by the absolute change since 2015
sql='''
SELECT Rank, 
SUM("2023" * "Absolute change since 2015") / SUM("Absolute change since 2015") AS weighted_avg_2023
FROM GHI
GROUP BY Rank
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Rank countries based on their improvement from 2015 to 2023 and return the top 10
sql='''
SELECT Country, ("2015" - "2023") AS improvement,
RANK() OVER (ORDER BY ("2015" - "2023") DESC) AS rank_improvement
FROM GHI
LIMIT 10
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Calculate the percentage difference in values between 2000 and 2023 for each country
sql='''
SELECT Country,
(("2023" - "2000") / "2000") * 100 AS pct_diff_2000_to_2023
FROM GHI
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Identify countries with a consistent decrease in values from 2000 to 2023
sql='''
SELECT Country FROM GHI
WHERE "2000" > "2008" AND "2008" > "2015" AND "2015" > "2023"
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Find the country with the highest average value across all years
sql='''
SELECT Country, 
(AVG("2000") + AVG("2008") + AVG("2015") + AVG("2023")) / 4 AS avg_value
FROM GHI
GROUP BY Country
ORDER BY avg_value DESC
LIMIT 1
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select the top 5 countries with the highest total decrease from 2000 to 2023
sql='''
SELECT Country, ("2000" - "2023") AS total_decrease
FROM GHI
ORDER BY total_decrease DESC
LIMIT 5
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```
